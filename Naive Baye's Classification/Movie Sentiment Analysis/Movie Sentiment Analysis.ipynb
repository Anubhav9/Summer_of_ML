```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
reg=RegexpTokenizer(r'\w+')
```

```python
def clean_review(text):
    text=text.lower()
    text=text.replace("<br /><br />","")
    
    new_token=reg.tokenize(text)
    sw=stopwords.words("english")
    new_words=[new_token for new_token in new_token if new_token not in sw]
    ps=PorterStemmer()
    final_words=[ps.stem(new_token) for new_token in new_words]
    clean_review_final=" ".join(final_words)
    return clean_review_final
```
