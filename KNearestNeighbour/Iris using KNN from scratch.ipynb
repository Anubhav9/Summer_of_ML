```python
def distance(x1,x2):
    
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(X,y,querypoint,k):
    values=[]
    array=np.zeros(3)
    top=0
    maxi=0
    
    m=X.shape[0]
    for i in range(m):
        
        t=distance(X[i],querypoint)
        values.append((t,y[i]))
    values=sorted(values)
    values=values[0:k]
    for i in range(k):
        if(values[i][1]==0):
            array[0]=array[0]+1
        if(values[i][1]==1):
            array[1]=array[1]+1
        if(values[i][1]==2):
            array[2]=array[2]+1
    
    for i in range(1):
        if(array[i]>array[i+1]):
            maxi=i
        else:
            maxi=i+1
        
        if(array[maxi]>array[i+2]):
            top=maxi
        else:
            top=i+2
        
   
    return top
        
```

```python
knn(X,y,(5.5,2.3,4,1.3),5)
```
