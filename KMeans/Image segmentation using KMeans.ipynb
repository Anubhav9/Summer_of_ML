```python
pixels=im.reshape((1080*1920,3))
```

```python
im.shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
km
```

```python
km.fit(pixels)
```

```python
cm=km.cluster_centers_
```

```python
km.labels_
```

```python
cm=np.array(cm,dtype=int)
```

```python
cm
```

```python
i=1
plt.figure(0,figsize=(4,2))
colors=[]
for each_color in cm:
    a=plt.subplot(1,3,i)
    print(each_color)
    
    i=i+1
    colors.append(each_color)
    b=np.zeros((100,100,3),dtype='uint8')
    b[:,:,:]=each_color
    plt.imshow(b)
    
```

```python
colors
```

```python
new_img=np.zeros((1080*1920,3),dtype='uint8')
```

```python
new_img.shape
```

```python
m=new_img.shape[0]
```

```python
m
```

```python
for i in range (m):
    new_img[i]=colors[km.labels_[i]]
new_img=new_img.reshape((1080,1920,3))
```
