```python
#fitting the data

X=x-(x.mean())/x.std()
Y=y
```

```python
#again plotting x and y

plt.scatter(X,Y)
```

```python
#now applying the linear regression algorithm

def hypothesis(thetha,x):
    return thetha[0]+thetha[1]*x

def error(X,Y,thetha):
    m=X.shape[0]
    total_error=0
    for i in range(0,m):
        hx=hypothesis(thetha,X[i])
        total_error=total_error+(Y[i]-hx)**2
    return total_error
def gradient(X,Y,thetha):
    grad=np.zeros((2,))
    m=X.shape[0]
    for i in range(m):
        hx=hypothesis(thetha,X[i])
        grad[0] +=(hx-Y[i])
        grad[1] +=(hx-Y[i])*X[i]
    return grad
def gradientDescent(X,Y,learning_rate=0.00001):
    thetha=np.zeros((2,))
    max_itr=800
    itr=0
    el=[]
    while(itr<max_itr):
        grad=gradient(X,Y,thetha)
        
        e=error(X,Y,thetha)
        el.append(e)
        thetha[0]=thetha[0]-learning_rate*grad[0]
        thetha[1]=thetha[1]-learning_rate*grad[1]
        itr=itr+1
    return thetha,el
        
    
    
```

```python
final_thetha,final_error=gradientDescent(X,y)
```
